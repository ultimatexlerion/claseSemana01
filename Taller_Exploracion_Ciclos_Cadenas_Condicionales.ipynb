# Taller: Exploración de Datos con Ciclos, Cadenas y Condicionales en Python

## Objetivo del Taller
- Aplicar el uso de ciclos, manejo de cadenas y condicionales para explorar y analizar datos en un archivo CSV utilizando Python.

## Paso 1: Preparar el Entorno de Trabajo
1. **Carga del Archivo CSV:**
- Asegúrate de haber subido el archivo CSV a Google Colab como se indicó en el taller anterior.
- Carga el archivo en un DataFrame de `pandas` utilizando el siguiente código:


```python
import pandas as pd

# Reemplaza 'nombre_del_archivo.csv' con el nombre real de tu archivo CSV
df = pd.read_csv('/content/tested.csv')
df.head()
```

## Paso 2: Exploración de Datos con Ciclos
1. **Iterar Sobre Filas:**
- Itera sobre las filas del DataFrame para explorar el contenido de cada fila. Utiliza un ciclo `for` para recorrer las filas:


```python
for index, row in df.iterrows():
    print(f'Age {index}: {row}')
df.head()
```

2. **Contar Valores con una Condición:**
- Usa un ciclo para contar cuántas filas cumplen con una condición específica. Por ejemplo, cuántos registros tienen una cadena específica en una columna.


```python
# Reemplaza 'Nombre_de_la_columna' y 'valor_especifico' por los valores que deseas analizar
contador = 0
for index, row in df.iterrows():
    if 'Kelly' in row['Name']:
        contador += 1

print(f'Total de filas que cumplen con la condición: {contador}')
```

## Paso 3: Manipulación de Cadenas
1. **Operaciones con Cadenas:**
- Manipula cadenas en las columnas para extraer información o realizar modificaciones. Por ejemplo, convertir una columna a minúsculas:


```python
# Reemplaza 'Nombre_de_la_columna' con el nombre de la columna
df['PassengerId'] = df['Name'].str.lower()
print(df.head())
```

2. **Filtrar Filas Basado en Cadenas:**
- Filtra las filas que contienen una cadena específica en una columna y muestra el resultado:


```python
# Reemplaza 'Nombre_de_la_columna' y 'cadena_a_buscar'
filtro = df[df['Name'].str.contains('kelly', case=False)]
print(filtro)
```

## Paso 4: Uso de Condicionales para Análisis
1. **Clasificación Basada en Condiciones:**
- Clasifica las filas según una condición y añade una nueva columna con el resultado de la clasificación:


```python
# Reemplaza 'Nombre_de_la_columna' y ajusta las condiciones a tus necesidades
condiciones = []
for index, row in df.iterrows():
    if row['Survived'] > 10:
        condiciones.append('Alto')
    else:
        condiciones.append('Bajo')

df['Clasificación'] = condiciones
print(df.head())
```

2. **Condicionales Múltiples:**
- Aplica múltiples condiciones para realizar un análisis más detallado. Por ejemplo, clasificar según dos columnas:


```python
def clasificar(row):
    if row['Ticket'] == 10 and row['Fare'] == 'Específico':
        return 'Clase A'
    else:
        return 'Clase B'

df['Clase'] = df.apply(clasificar, axis=1)
print(df.head())
```
