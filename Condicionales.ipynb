Los operadores son simbolos especiales que representan cálculos, como la suma o la multiplicación. Los valores a los cuales se aplican esos operadores reciben el nombre de operandos. Los principales operadores sobre los tipo int y float.



Operadores  | correspondencia|ejemplo|
  ---   | ---|---
