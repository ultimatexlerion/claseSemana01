# Taller: Descarga Manual y Análisis de Datos CSV en Google Colab

## Objetivo del Taller
- Aprender a descargar un archivo CSV desde un portal de datos abiertos, cargarlo en Google Colab y realizar un análisis básico utilizando Python.

## Paso 1: Descarga Manual del Archivo CSV
1. **Accede al Portal de Datos Abiertos:**
   - Visita la [página del dataset](https://www.datos.gov.co/Organismos-de-Control/Registro-de-Activos-de-Informaci-n-Personer-a-Muni/b2he-ipst/about_data).
2. **Exportar el Archivo CSV:**
   - En la página del dataset, localiza el botón `Exportar` y selecciónalo.
   - Selecciona la opción `CSV` como formato de exportación.
   - Haz clic en `Descargar` para guardar el archivo en tu computadora.

## Paso 2: Subir el Archivo CSV a Google Colab
1. **Accede a Google Colab:**
   - Abre [Google Colab](https://colab.research.google.com/) en tu navegador.
2. **Subir el Archivo CSV:**
   - En la barra lateral de Google Colab, haz clic en el ícono de carpeta para abrir el explorador de archivos.
   - Haz clic en el botón `Subir` y selecciona el archivo CSV que descargaste previamente en tu computadora.
